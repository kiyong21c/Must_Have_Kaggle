#### 피처 요약표 만들기

````
# 피처 요약표 만들기

def resumetable(df):
    print(f'데이터 형상 : {df.shape}')
    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입']) # 인덱스 : 피처명, 값 : dtypes
    summary = summary.reset_index() # 기존 인덱스인 피처명이 컬럼으로 들어감(컬럼이름은 'index')
    summary = summary.rename(columns={'index':'피처'}) # 컬럼명 rename : 'index' → '피처'
    summary['결측값 개수'] = df.isnull().sum().values # 결측값이면 True → 컬럼별 sum → 배열
    summary['고유값 개수'] = df.nunique().values
    summary['첫 번째 값'] = df.loc[0].values
    summary['두 번째 값'] = df.loc[1].values
    summary['세 번째 값'] = df.loc[2].values

    return summary
````

````
# 비율 텍스트를 표시하는 함수
def write_percent(ax, total_size):
    for patch in ax.patches:
        height = patch.get_height()
        width = patch.get_width()
        left_coord = patch.get_x()
        percent = height/total_size * 100

        ax.text(x=left_coord + width/2.0,       # x축 위치
                y=height + total_size*0.001,    # y축 위치
                s=f'{percent:1.1f}%',           # 입력 텍스트
                ha='center')                    # 정렬방식
````

#### 1.교차분석표 생성 함수 만들기
 - 범주형 데이터 2개를 비교 분석하는데 사용되는 표
 - 목적 : 명목형 피처(=범주형 데이터)별 타깃값 '1' 비율확인
 - 판다스의 crosstab() 함수 : pd.crosstab(피처1, 피처2)

 - 상기의 포인트 플롯을 그리기 위해 교차분석표 만듬

````
crosstab = pd.crosstab(train['nom_0'], train['target']) # 'nom_0' 범주형 데이터 각각에 대한 'target' 값 각각의 수
crosstab
````

 - 개수보다는 비율로 표현하는것이 의미파악 용이함
 - normalize파라미터에 'index' 전달 → 인덱스 기준으로 정규화

````
crosstab = pd.crosstab(train['nom_0'], train['target'], normalize='index') * 100
crosstab
````

 - 현재 인덱스가 피처 이름(nom_0)
 - 이를 열로 가져와야 그래프를 그리기 편하다

````
crosstab = crosstab.reset_index()
crosstab
````

 - 상기의 과정들을 계속사용 하기위에서 함수로 제작

````
def get_crosstab(df, feature):
    crosstab = pd.crosstab(df[feature], df['target'], normalize='index') * 100
    crosstab = crosstab.reset_index()
    return crosstab
````

 - 포인트플롯에서는 타깃값 1에 대한 비율만 나타낼 것이므로 교차분석표에서 타깃값 1만 가져옴

#### 2.포인트플롯 생성 함수 만들기
 - 세가지를 파라미터로 받자

  > ax : 포인트플롯을 그릴 축

  > feature : 포인트플롯으로 그릴 피처

  > crosstab : 교차분석표

````
# 기존 축에다가 x축을 공유하는 새로운 축을 만들면서 포인트플롯 생성
def plot_pointplot(ax, feature, crosstab):
    ax2 = ax.twinx() # ax와 x축만 공유하는 새로운 ax2축 생성
    ax2 = sns.pointplot(x=feature, y=1, data=crosstab,  # sns.pointplot() → return AxesSubplot
                        order=crosstab[feature].values,
                        color='black',
                        legend=False)
    ax2.set_ylim(crosstab[1].min()-5, crosstab[1].max()*1.1)
    ax2.set_ylabel('Target Ration(%)')
````

#### 3.피처 분포도 및 피처별 타깃값 1의 비율 포인트플롯 생성 함수 만들기
 - 피처 분포도 생성 함수 + 포인트플롯 생성함수

````
def plot_cat_dist_with_true_ratio(df, features, num_rows, num_cols, size=(15, 20)):
    plt.figure(figsize=size) # 전체 Figure 크기 설정
    grid = gridspec.GridSpec(num_rows, num_cols) # 그리드스펙 객체 생성
    plt.subplots_adjust(wspace=0.45, hspace=0.3)

    for idx, feature in enumerate(features):
        ax = plt.subplot(grid[idx]) # 피처순서대로 그리드스펙객체에 할당하여 축 생성
        crosstab = get_crosstab(df, feature) # 교차분석표 생성 함수

        # ax축에 타깃값 분포 카운트플롯 그리기
        sns.countplot(x=feature, data=df,
                      order=crosstab[feature].values,
                      color='skyblue',
                      ax=ax)
        write_percent(ax, len(df)) # 비율표시 함수

        # ax2축에 포인트플롯 그리기
        plot_pointplot(ax, feature, crosstab)

        # 축 제목 설정
        ax.set_title(f'{feature} Distribution') 
````

````
nom_features = ['nom_0','nom_1','nom_2','nom_3','nom_4'] # 명목형 피처
plot_cat_dist_with_true_ratio(train, nom_features, num_rows=3, num_cols=2)
````

  > nom_4피처의 각 값인 'Bassoon', 'Piano', 'Theremin', 'Oboe' 개수의 비율 → 막대그래프

  > nom_4피처의 값 중 'Theremin'인 경우 타깃값이 1인경우가 가장 높으며, 36% 정도이다 → 포인트플롯그래프

````
ord_features = ['ord_0','ord_1','ord_2','ord_3'] # 명목형 피처
plot_cat_dist_with_true_ratio(train, ord_features, num_rows=2, num_cols=2, size=(15, 12))
````

 - ord_1, ord_2 는 피처값들의 순서가 뒤죽박죽

  > ord_1 : 'Novice','Contributior','Expert','Master','Grandmaster' 순으로 정렬

  > ord_2 : 'Freezing','Cold','Warm','Hot','Boiling Hot','Lava Hot' 순으로 정렬

 - CategoricalDtype()을 이용하여 피처에 각 범주형 데이터에 순서를 지정할 수 있음

 - countplot()함수를 만들때 order파라미터에 순서를 집어넣어도 되지만 그래프상에서만 해당 순서로 보일뿐 원래 데이터 자체에 순서가 생기는 것이 아님

  > 기존 train['ord_1']의 dtype : object

  > 변경 train['ord_1']의 dtype : category(순서가 지정된)

````
train['ord_1']
````

````
from pandas.api.types import CategoricalDtype

ord_1_value = ['Novice','Contributior','Expert','Master','Grandmaster']
ord_2_value = ['Freezing','Cold','Warm','Hot','Boiling Hot','Lava Hot']

# 순서를 지정한 범주형 데이터 타입 객체 생성
ord_1_dtype = CategoricalDtype(categories=ord_1_value, ordered=True)
ord_2_dtype = CategoricalDtype(categories=ord_2_value, ordered=True)

# 데이터 타입 변경(object → CategoricalDtype)
train['ord_1'] = train['ord_1'].astype(ord_1_dtype)
train['ord_2'] = train['ord_2'].astype(ord_2_dtype)
````
